```python
# 📌 Importations
import os
import fitz  # PyMuPDF
import faiss
import numpy as np
import ipywidgets as widgets
from sentence_transformers import SentenceTransformer
from IPython.display import display
import ollama  # Assurez-vous que Ollama est bien configuré
```

```python
MODEL = "minicpm-v"  # 🔹 Changez selon le modèle que vous utilisez

# 📌 Initialisation du modèle d'embedding
embedder = SentenceTransformer('all-MiniLM-L6-v2')
```

```python
PDF_PATH = "/content/investigatePHASE.pdf"  # 🟢 Modifier ce chemin si besoin
```

```python
# 📌 Fonction pour extraire le texte du PDF
def extract_pdf_text(pdf_path):
    document = fitz.open(pdf_path)  # Ouvrir le PDF
    text = ""
    for page_num in range(document.page_count):
        page = document.load_page(page_num)
        text += page.get_text("text")  # Extraire le texte de chaque page
    return text

# 📌 Fonction pour générer l'embedding d'une requête
def generate_query_embedding(query):
    return embedder.encode(query, convert_to_tensor=True)

# 📌 Fonction pour découper le texte en morceaux
def chunk_text(text, chunk_size=500):
    words = text.split()
    chunks = [" ".join(words[i:i + chunk_size]) for i in range(0, len(words), chunk_size)]
    return chunks

# 📌 Fonction pour générer les embeddings des chunks
def build_embeddings(chunks):
    return embedder.encode(chunks, convert_to_tensor=True)

# 📌 Stockage des embeddings dans FAISS
def store_embeddings_in_faiss(embeddings):
    embeddings_cpu = embeddings.cpu().detach().numpy()
    dimension = embeddings_cpu.shape[1]
    index = faiss.IndexFlatL2(dimension)
    index.add(np.ascontiguousarray(embeddings_cpu, dtype='float32'))
    return index

# 📌 Recherche du contexte pertinent
def search_relevant_context(query, index, chunks):
    query_embedding = generate_query_embedding(query)
    query_embedding_cpu = query_embedding.cpu().detach().numpy().reshape(1, -1)
    D, I = index.search(query_embedding_cpu, k=1)
    return chunks[I[0][0]]

# 📌 Construction et stockage de l'index FAISS
def build_and_store_index(pdf_path):
    pdf_text = extract_pdf_text(pdf_path)
    chunks = chunk_text(pdf_text)
    embeddings = build_embeddings(chunks)
    index = store_embeddings_in_faiss(embeddings)
    return index, chunks

# 📌 Génération du prompt avec contexte
def build_prompt_with_context(query, index, chunks):
    relevant_context = search_relevant_context(query, index, chunks)

    messages = [
        {"role": "user", "content": (
            "You are an expert AI assistant that will answer strictly based on the provided context. "
            "Do not use external knowledge. Here is the relevant context from the PDF:\n\n"
            f"{relevant_context}\n\n"
            "Now, answer the following query:\n"
        )},
        {"role": "user", "content": query}
    ]

    return {"model": MODEL, "messages": messages}

# 📌 Chargement et indexation du PDF
if not os.path.exists(PDF_PATH):
    print(f"⚠️ Erreur : Le fichier PDF '{PDF_PATH}' est introuvable.")
else:
    index, chunks = build_and_store_index(PDF_PATH)
    print("✅ Indexation terminée ! Vous pouvez poser vos questions.")

# 📌 Interface utilisateur pour poser des questions
query_input = widgets.Text(
    placeholder="Entrez votre question ici...",
    layout=widgets.Layout(width='70%')
)
submit_button = widgets.Button(description="Poser la question")
output_area = widgets.Output()

def on_submit(b):
    with output_area:
        output_area.clear_output()
        query = query_input.value.strip()
        if query:
            prompt = build_prompt_with_context(query, index, chunks)
            try:
                response = ollama.chat(**prompt)
                print("\n🔹 Réponse de l'IA :")
                print(response['message']['content'])
            except Exception as e:
                print(f"⚠️ Erreur de communication avec Ollama : {e}")

submit_button.on_click(on_submit)

display(query_input, submit_button, output_area)
```
